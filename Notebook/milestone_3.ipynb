```python
w = torch.tensor([0.5], requires_grad=True)

# Manually set gradient
w.grad = torch.tensor([0.2])

# SGD update
learning_rate = 0.01
with torch.no_grad():
    w -= learning_rate * w.grad

print("Updated weight:", w.item())
```
